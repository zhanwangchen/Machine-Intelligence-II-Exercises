```python
 %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
import skimage 
```

```python
img = np.invert(misc.imread ('testimg.jpg'))
plt.figure()
plt.imshow(img, cmap='binary')
plt.show()
vector = np.reshape(img,177500)
# Normalization of the data
def MaxMinNormalization(x,Max,Min): 
    x = (x - Min)/(Max - Min)
    return x 
vector_normalized= MaxMinNormalization(vector,np.max(vector),np.min(vector))
#plot the original database

x = np.linspace(0,10,177500)
y = vector_normalized
plt.figure(figsize=(20,10))
plt.plot(x, y)
plt.title('Original Dataset without noises',fontsize=20)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.show() 
```
