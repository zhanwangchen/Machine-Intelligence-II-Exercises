```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from pandas.tools.plotting import scatter_matrix
from numpy import *
```

```python
#Task 4.1(a)
#Create a toy dataset of 2-dimensional data points
#Sampling from normal distribution of given mean value and standard deviation 0.1
xa=np.ones([90,2])
xa[0:30,0]=np.random.normal(-0.5,0.1,[30])
xa[0:30,1]=np.random.normal(-0.2,0.1,[30])
xa[30:60,0]=np.random.normal(0,0.1,[30])
xa[30:60,1]=np.random.normal(0.6,0.1,[30])
xa[60:90,0]=np.random.normal(0.5,0.1,[30])
xa[60:90,1]=np.random.normal(0,0.1,[30])
#plot the data points from 3 different colors
plt.figure(figsize=(8,8))
plt.scatter(xa[0:30,0],xa[0:30,1],label='first sample(-0.5,-0.2)',color='green') 
plt.scatter(xa[30:60,0],xa[30:60,1],label='second sample(0,0.6)',color='red') 
plt.scatter(xa[60:90,0],xa[60:90,1],label='third sample(0.5,0)',color='blue') 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of the generated toy dataset',fontsize=30)
plt.legend()
plt.show()
```

```python
#Task 4.1(b)
#Using Kernel trick
#define RBF kernel ,  gamma=1/(2*width^2)
from scipy.spatial.distance import pdist, squareform

def rbf_kpca(X, gamma, k):
    sq_dist = pdist(X, metric='sqeuclidean')
                            # N = X.shape[0]    
                            # sq_dist.shape = N*(N-1)/2
    mat_sq_dist = squareform(sq_dist)
                            # mat_sq_dist.shape = (N, N)
    # step 1
    K = np.exp(-gamma*mat_sq_dist)

    # step 2
    N = X.shape[0]
    one_N = np.ones((N, N))/N
    K = K - one_N.dot(K) - K.dot(one_N) + one_N.dot(K).dot(one_N)

    # step 3
    Lambda, Q = np.linalg.eig(K)
    eigen_pairs = [(Lambda[i], Q[:, i]) for i in range(len(Lambda))]
    eigen_pairs = sorted(eigen_pairs, reverse=True, key=lambda k: k[0])
    return np.column_stack((eigen_pairs[i][1] for i in range(k)))
```

```python
#Choosing width, then  get a suitable gamma
xa_kpca = rbf_kpca(xa, gamma=4, k=2)
```

```python
#plotting the transformed data points
plt.figure(figsize=(8,8))
plt.scatter(xa_kpca[0:30,0],xa_kpca[0:30,1],color='green') 
plt.scatter(xa_kpca[30:60,0],xa_kpca[30:60,1],color='red') 
plt.scatter(xa_kpca[60:90,0],xa_kpca[60:90,1],color='blue') 
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
